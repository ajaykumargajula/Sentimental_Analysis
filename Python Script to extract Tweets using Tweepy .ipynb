#### Authentication for Twitter Login

```python
import tweepy
import json

consumer_key="xxxxxxxxxxxxxxxxxx"
consumer_secret_key="xxxxxxxxxxxxxxxxxx"
access_key="xxxxxxxxxxxxxxxxxx"
access_secret_key="xxxxxxxxxxxxxxxxxx"

auth=tweepy.OAuthHandler(consumer_key,consumer_secret_key)
auth.set_access_token(access_key,access_secret_key)
api=tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)

try:
    api.verify_credentials()
    print("Authentication is Successful")
except:
    print("Authentication Failed.. Please check your Keys")
    
```

#### Get Popular Trends in India

```python
trend_result=api.trends_place(2282863)
for trend in trend_result[0]["trends"]:
    print(trend["name"])
```

### Extact Tweets for query on Hashtag from popular Trends

```python
Tweets=[]
new_tweets=api.search(q="#PresidentRule",lang="en",count=100,tweet_mode="extended")
Tweets.extend(new_tweets)

last_id=Tweets[-1].id-1

while len(new_tweets)>0:
    print("Extracting Tweets before id {}".format(last_id))
    new_tweets=api.search(q="#PresidentRule",lang="en",count=100,max_id=last_id,tweet_mode="extended")
    Tweets.extend(new_tweets)
    last_id=Tweets[-1].id-1
    print("Tweets Extracted {}".format(len(Tweets)))

tweets_output=[[tweet.id_str, tweet.created_at, tweet.full_text.encode("utf-8")] for tweet in Tweets]
```
